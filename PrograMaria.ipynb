## Distribuição amostral e intervalo de confiança

```python
salario = planilha["SALARIO"]
salario
```

```python
media_amostral = np.mean(salario)
media_amostral
```

```python
desvio_amostral = np.std(salario)
desvio_amostral
```

```python
nivel_confianca = 0.95
```

```python
tamanho_amostra = len(salario)
tamanho_amostra
```

```python
from scipy import stats
```

```python
erro_padrao = stats.sem(salario)
erro_padrao
```

```python
intervalo_confianca = stats.t.interval(nivel_confianca, tamanho_amostra - 1, loc=media_amostral, scale=erro_padrao)
intervalo_confianca
```

### Desafio - Calcular o nível de confiança de (90%) por cargo e por gênero.

```python
salario_feminino = planilha[planilha["GENERO"] == "Feminino"]
grupos_cargo = salario_feminino.groupby("CARGO ATUAL")["SALARIO"]
grupos_cargo

grupos_cargo_filtrado = grupos_cargo.filter(lambda x: len(x) > 1)
```

```python
for cargo, grupo_salario in grupos_cargo:
    print("Cargo:", cargo)
    print(grupo_salario)
    print("-" * 30)
```

```python
media_grupo_genero = grupos_cargo_filtrado.mean()
desvio_grupo_genero = grupos_cargo_filtrado.std()
nivel_confianca = 0.9
tamanho_amostra_grupo = len(grupos_cargo_filtrado)
erro_padrao = grupos_cargo_filtrado.sem()
```

```python
intervalo_confianca = stats.t.interval(nivel_confianca, tamanho_amostra_grupo - 1, loc=media_grupo_genero, scale=erro_padrao)
intervalo_confianca
```
