# What's the most expensive wine?

```python
max_price = wines[wines["price"] == wines["price"].max()]
max_price
```

```python
max_price_name = max_price["name"].iloc[0]
max_price_name
```

# Calculate a new column where you show the ratings in a scale from 0 to 5. Floats are allowed 

```python
wines["NEW_RATING"] = wines["points"] / 100 * 5
```

```python
wines
```

# Create a price histogram for wines that cost less than 100

```python
low_price = wines[wines["price"] < 100]
low_price
```

```python
low_price["price"].hist()
```

# Plot size horizontally against points vertically. Do you think ratings increase by price? 

```python
wines.plot(x="price", y="points", figsize=(15,3), kind=("scatter"))
```
